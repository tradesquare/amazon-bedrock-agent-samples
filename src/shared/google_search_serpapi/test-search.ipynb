```python
from serpapi import GoogleSearch
import dotenv
import os
dotenv.load_dotenv()

query = '''‡∏Å‡∏°‡∏•‡πÇ‡∏•‡∏´‡∏∞‡∏Å‡∏¥‡∏à'''
api_key = os.getenv("SERPAPI_KEY")

def _search(query, params):
    params["q"] = query
    search = GoogleSearch(params)
    results = search.get_dict()
    return results
```

```python
query = '''‡∏Å‡∏°‡∏•‡πÇ‡∏•‡∏´‡∏∞‡∏Å‡∏¥‡∏à'''
params = {
  "location": "Bangkok, Thailand",\
  "hl": "en",
  "gl": "th",
  "google_domain": "google.co.th",
  "api_key": api_key
}

results = _search(query, params)
results
```

```python
from urllib.parse import urlparse
import re

compnay_ref_web = ['data.creden.co', 'dbd.go.th']
duplciated_company = []

# get similar company name from DBD or CREDEN
for res in results['inline_images']:
    # print(res['title'])
    # print(res['source'])
    # print(res['source_name'])
    # print(urlparse(res['source']).netloc)
    if urlparse(res['source']).netloc in compnay_ref_web:
        print(res)
        # matched_name = re.match(query)[0]
        # print(matched_name)
        duplciated_company.append(res['title'])
```

```python
import re

company_names = []
for comp_name in duplciated_company:
    match = [re.search(f"{query}.+",chunk) for chunk in comp_name.split(' ')]
    name = [m.group(0) for m in match if m]
    company_names.extend(name)
```
